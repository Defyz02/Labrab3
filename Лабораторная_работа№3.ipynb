```python
#считываем данные с интернет ресурса
def create_url(date):
  url_begin = "https://so-ups.ru/functioning/ees/ees-indicators/ees-gen-consump-hour/?tx_mscdugraph_pi%5Bcontroller%5D=Graph&tx_mscdugraph_pi%5Baction%5D=fullview&tx_mscdugraph_pi%5BviewDate%5D="
  url_end = "&tx_mscdugraph_pi%5BviewKpo%5D=6100"
  return url_begin + date + url_end

def parse_consumption(date):
  url = create_url(date)
  soup = BeautifulSoup(requests.get(url).content, 'html.parser')
  full_txt = str(soup)
  text_list_datetime = full_txt[full_txt.find('data-datax=') + len('data-datax=') + 1 : full_txt.find('data-datay=') - 2].split(',')
  text_list_consumption = full_txt[full_txt.find('data-datay=') + len('data-datay=') + 1 : full_txt.find('data-datay1=') - 2].split(',')
  text_list_generation = full_txt[full_txt.find('data-datay1=') + len('data-datay1=') + 1 : full_txt.find('data-date=') - 2].split(',')
  df = pd.DataFrame([text_list_datetime, list(map(int, text_list_generation)), list(map(int, text_list_consumption))])
  return df.T
```

```python
# берем данные за некотоырй промежуток и группируем их в одну таблицу
df_gen_4 = parse_consumption('2022-01-01')
df_gen_4 = df_gen_4.append(parse_consumption('2022-01-20'))
df_gen_4 = df_gen_4.append(parse_consumption('2022-02-01'))
df_gen_4
```

```python
# . Переименуем столбцы
df_gen_5 = df_gen_4.copy()

df_gen_5.columns=['Время', 'Генерация (МВт)', 'Потребление (МВт)']
df_gen_5=df_gen_5.set_index('Время')
df_gen_5['Генерация (МВт)']=df_gen_5['Генерация (МВт)'].astype('int')
df_gen_5['Потребление (МВт)']=df_gen_5['Потребление (МВт)'].astype('int')
df_gen_5['Разность (МВт)']=df_gen_5['Генерация (МВт)']-df_gen_5['Потребление (МВт)'] 

df_gen_5
```

```python
axes = df_gen_5.plot(marker='.', figsize=(20, 10), subplots = True)

for _ in axes:
  _.grid(True)

axes = df_gen_5[df_gen_5.columns[: -1]].plot(marker='.', figsize=(20, 5), subplots = False)
axes.grid(True)
```

```python
df_ex = pd.read_excel('https://drive.google.com/uc?export=download&id=1IFULWv72kGpZRDRpqRskm5NIM6AOCFPR', header = None)
df_ex
```

```python
#разделяем столбец время 
df_day = df_ex[0].str.split('.', expand = True) 
print(df_day.head())
print()

# создаем изначально пустую таблицу и заполняем нужными нам значениями
df_gen_6 = pd.DataFrame()

df_gen_6['Генерация (МВт)'] = df_ex[1]
df_gen_6['Месяц'] = df_day[1].astype(int) 

print(df_gen_6.head())
```

```python
#получаем статистику за год
day_stat = df_gen_6.groupby(['Месяц'])['Генерация (МВт)'].sum()

print(day_stat)
print()

day_stat.plot(kind = 'bar')
print()
```

```python
#круговая диаграмма потребления по месяцам
gen = df_gen_6.groupby(['Месяц']).sum()


labels = gen.index

fig1, ax1 = plt.subplots(figsize = (10, 10))
ax1.pie(gen['Генерация (МВт)'] ,labels=gen.index, autopct = '%1.0f%%')
ax1.axis('equal')  

plt.show()
```
